### Making a list of non-synnonymous mutations for the plot from the annotated allele

```python
dfsolar = pd.read_csv('result_soil_nutrient_retention.csv')
dfsolar.head()

m = []
l = []

for j in range(len(dfsolar)):
    if dfsolar['Significant SNPs'][j] not in m:
        if dfsolar['Annotation'][j] == 'missense_variant':
            l.append((dfsolar['Significant SNPs'][j],dfsolar['Gene Function'][j]))
            m.append(dfsolar['Significant SNPs'][j])

print(len(l))
l
```

### If list contains more than one allele, plot with the following code

```python
#Solar radiation
fig, axes = plt.subplots(110,figsize=(40,400))
fig.suptitle('Allele Frequency plot of soil_nutrient_retention genes')
#axes.set_ylim([0,1])
for i in range(110):
    axes[i].plot(col,dfNew.loc[l[i][0]],linewidth=3)
    #axes[i].plot(sorted(lon),dflon.loc[l[i][0]],linewidth=3)
    #axes[i].set(ylabel=s[i])
    #axes[i].set(ylabel=l[i][1].split(' / ')[-1])
    axes[i].set(ylabel=l[i][0])
    axes[i].set(xlabel=l[i][1])
    axes[i].set_ylim([0,1])

plt.savefig('AF-soil_nutrient_retention.pdf',dpi=1200)
plt.show()
```
